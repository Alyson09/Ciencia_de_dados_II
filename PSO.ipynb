```python
import random

def calculate_total_distance(route, distance_matrix):
    total_distance = 0
    for i in range(len(route)):
        total_distance += distance_matrix[route[i-1], route[i]]
    return total_distance

def create_solution(num_cities):
    solution = list(range(num_cities))
    random.shuffle(solution)
    return solution

def swap(route):
    new_route = route.copy()
    i, j = random.sample(range(len(route)), 2)
    new_route[i], new_route[j] = new_route[j], new_route[i]
    return new_route

def pso_tsp(distance_matrix, num_particles, max_iter):
    num_cities = len(distance_matrix)
    
    particles = [create_solution(num_cities) for _ in range(num_particles)]
    velocities = [random.sample(range(num_cities), 2) for _ in range(num_particles)]
    
    pbest = particles[:]
    gbest = min(particles, key=lambda p: calculate_total_distance(p, distance_matrix))
    
    for iteration in range(max_iter):
        for i in range(num_particles):
            if calculate_total_distance(particles[i], distance_matrix) < calculate_total_distance(pbest[i], distance_matrix):
                pbest[i] = particles[i]
            if calculate_total_distance(pbest[i], distance_matrix) < calculate_total_distance(gbest, distance_matrix):
                gbest = pbest[i]
                
            particles[i] = swap(particles[i])
            
    return gbest, calculate_total_distance(gbest, distance_matrix)

num_particles = 10
max_iter = 100

best_route, best_distance = pso_tsp(distance_matrix, num_particles, max_iter)

print("Melhor Rota:", best_route)
print("Menor Distância:", best_distance)
```
